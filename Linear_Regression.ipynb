```python
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Preprocessing phase
dataset = pd.read_csv("Salary_dataset.csv")
X = dataset.iloc[:, 1].values.reshape(-1, 1)
y = dataset.iloc[:, -1].values
```

```python
#Split the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
#Train the model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
#Predict
y_pred = regressor.predict(X_test)
#Compare predicted and actual
print(np.round(np.concatenate((y_pred.reshape(-1, 1), y_test.reshape(-1, 1)), axis=1), decimals=1))
```

```python
#Visualizing the Regression Line (Train set)
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_train, regressor.predict(X_train), color='red', label='Regression Line')
plt.title('Salary vs Experience (Training set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
#Visualizing the Regression Line (Test set)
plt.scatter(X_test, y_test, color='blue', label='Testing Data')
plt.plot(X_test, y_pred, color='red', label='Regression Line')
plt.title('Salary vs Experience (Test set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
